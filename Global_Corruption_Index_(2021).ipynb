**Data Visualization**

**IPK 2021 Secara Global**

```python
#menampilkan Top 10 Negara Secara Global dengan IPK Tertinggi 2021
pd.pivot_table(
    data=corruption_data_global,
    index='Country / Territory',
    values='CPI score 2021').round(2).sort_values(by ='CPI score 2021', ascending=False).head(10).reset_index()
```

```python
#menampilkan bentuk Visualisasi Data untuk Top 10 Negara Secara Global dengan IPK Tertinggi 2021
plt.figure(figsize=(12,5))
plt.title('Countries with Highest CPI Score (Perceived as Less Corupt)')
plt.ylabel('Corruption Perception Index (CPI) in 2021')
corruption_data_global.groupby('Country / Territory')['CPI score 2021'].sum().sort_values(ascending=False)[0:10].plot(kind='line', marker='*', color='red', ms=10)
corruption_data_global.groupby('Country / Territory')['CPI score 2021'].sum().sort_values(ascending=False)[0:10].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

- Berdasarkan data diatas, diperoleh bahwa :
  - **Denmark**, **Finlandia**, dan **Selandia Baru** merupakan negara dengan IPK tertinggi pada tahun 2021 dengan skor 88, disusul dengan **Norwegia** dan **Swedia**, dan **Singapura** dengan skor 85.
  - **Singapura** merupakan satu - satunya negara dari kawasan Asia Tenggara yang masuk ke jajaran 10 besar negara yang paling tidak korup di dunia tahun 2021.
  - 10 besar negara dengan angka IPK 2021 tertinggi tersebut dikategorikan sebagai negara yang paling tidak korup di dunia.

  

```python
#menampilkan Top 15 Negara Secara Global dengan IPK Terendah 2021
pd.pivot_table(
    data=corruption_data_global,
    index='Country / Territory',
    values='CPI score 2021').round(2).sort_values(by ='CPI score 2021', ascending=True).head(15).reset_index()
```

```python
#menampilkan bentuk Visualisasi Data untuk Top 15 Negara Secara Global dengan IPK Terendah 2021
plt.figure(figsize=(12,5))
plt.title('Countries with Lowest CPI Score (Perceived as More Corupt)')
plt.ylabel('Corruption Perceived Index (CPI) in 2021')
corruption_data_global.groupby('Country / Territory')['CPI score 2021'].sum().sort_values(ascending=True)[0:15].plot(kind='line', marker='*', color='red', ms=10)
corruption_data_global.groupby('Country / Territory')['CPI score 2021'].sum().sort_values(ascending=True)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

- Berdasarkan data diatas, diperoleh bahwa :
  - **South Sudan** merupakan negara dengan IPK terendah pada tahun 2021 dengan skor 11, disusul dengan **Syria**, **Somalia**, **Venezuela**, **Afghanistan**, **Korea Utara** dan **Yemen** dengan masing - masing skor IPK diantara 13 - 16. 
  - 15 besar negara dengan angka IPK 2021 terendah tersebut dikategorikan sebagai negara yang paling korup di dunia.
  

```python
#menampilkan IPK 2021 berdasarkan skala wilayah
pd.pivot_table(
    data=corruption_data_global,
    index='Region',
    values='CPI score 2021',
    aggfunc='sum').round(2).sort_values(by ='CPI score 2021', ascending=False).reset_index()
```

```python
#menampilkan bentuk Visualisasi Data untuk IPK 2021 berdasarkan skala wilayah
plt.figure(figsize=(12,5))
plt.title('CPI Score Based on Region')
plt.ylabel('Corruption Perceived Index (CPI) in 2021')
corruption_data_global.groupby('Region')['CPI score 2021'].sum().sort_values(ascending=False)[0:6].plot(kind='line', marker='*', color='red', ms=10)
corruption_data_global.groupby('Region')['CPI score 2021'].sum().sort_values(ascending=False)[0:6].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

```python
#menampilkan IPK 2021 tertinggi negara dari wilayah WE/EU 
pd.pivot_table(
    data=corruption_data_global[corruption_data_global["Region"]=="WE/EU"],
    index='Country / Territory',
    values='CPI score 2021').round(2).sort_values(by = 'CPI score 2021', ascending=False).head(5).reset_index()
```

```python
#menampilkan IPK 2021 terendah negara dari wilayah ECA
pd.pivot_table(
    data=corruption_data_global[corruption_data_global["Region"]=="ECA"],
    index='Country / Territory',
    values='CPI score 2021').round(2).sort_values(by = 'CPI score 2021', ascending=True).head(5).reset_index()
```

- Berdasarkan hasil akumulasi IPK 2021 per wilayah menunjukkan bahwa 
 - Wilayah WE/EU (European Union) memiliki skor IPK Tertinggi (Perceived as Less Corupt) jika dibandingkan dengan wilayah lainnya. Beberapa negara yang berasal dari wilayah ini antara lain negara Denmark, Finlandia, Norwegia, Swiss, dan Swedia.
  - Wilayah ECA (Europe & Central Asia) memiliki skor IPK Terendah (Perceived as More Corrupt) jika dibandingkan dengan wilayah lainnya. Beberapa contoh negara yang berasal dari wilayah ini adalah Turkmenistan, Uzbekistan, Tajikistan, Kyrgystan, dan Russia.

 

**IPK 2021 di Negara Bagian Asia & Pasifik**

```python
#menampilkan Top 5 Negara Bagian Asia & Pasifik dengan IPK Tertinggi 2021
pd.pivot_table(
    data=corruption_data_asia,
    index='Country / Territory',
    values='CPI score 2021').round(2).sort_values(by ='CPI score 2021', ascending=False).head(10).reset_index()
```

- Berdasarkan data diatas, diperoleh bahwa :
  - New Zealand merupakan negara dengan IPK tertinggi pada tahun 2021 dikawasan Asia & Pasifik dengan skor 88, disusul dengan negara Singapura (85), Hong Kong (76), Jepang (73), dan Australia (73). 
 - 15 besar negara dengan angka IPK 2021 tertinggi tersebut dikategorikan sebagai negara yang paling tidak korup di kawasan Asia & Pasifik.
 - Singapura dan Malaysia merupakan negara dari kawasan Asia Tenggara yang termasuk ke dalam 15 besar negara paling tidak korup di kawasan Asia dan Pasifik tahun 2021.

```python
#menampilkan Top 15 Negara Kawasan Asia dan Pasifik dengan IPK Terendah 2021
pd.pivot_table(
    data=corruption_data_asia,
    index='Country / Territory',
    values='CPI score 2021').round(2).sort_values(by ='CPI score 2021', ascending=True).head(15).reset_index()
```

```python
#menampilkan bentuk Visualisasi Data untuk Top 15 Negara Kawasan Asia & Pasifik dengan IPK Terendah 2021
plt.figure(figsize=(12,5))
plt.title('Countries in Asia & The Pasific with Lowest CPI Score (Perceived as More Corupt)')
plt.ylabel('Corruption Perceived Index (CPI) in 2021')
corruption_data_asia.groupby('Country / Territory')['CPI score 2021'].sum().sort_values(ascending=True)[0:15].plot(kind='line', marker='*', color='red', ms=10)
corruption_data_asia.groupby('Country / Territory')['CPI score 2021'].sum().sort_values(ascending=True)[0:15].plot(kind='bar',color=sns.color_palette("inferno_r", 7))
plt.show()
```

- Berdasarkan data diatas, diperoleh bahwa :
  - Afghanistan dan Korea Utara merupakan negara dengan IPK terendah pada tahun 2021 dengan skor 16, disusul dengan negara Kamboja (23), Bangladesh (26), Pakistan (28), dan Myanmar (28).
 - 15 besar negara dengan angka IPK 2021 terendah tersebut dikategorikan sebagai negara yang paling korup di kawasan Asia & Pasifik.
 -  Indonesia merupakan salah satu negara dari kawasan Asia Tenggara yang termasuk ke dalam negara paling korup di kawasan Asia & Pasifik tahun 2021.

**Ranking & Skor IPK 2021 Negara Indonesia**

```python
#menampilkan ranking dan IPK 2021 yang diperoleh negara Indonesia

pd.pivot_table(
    data=corruption_data_asia[corruption_data_asia["Country / Territory"]=="Indonesia"],
    index='Rank',
    values='CPI score 2021').round(2).sort_values(by = 'CPI score 2021', ascending=False)
```

  - Pada kawasan Asia Tenggara, Indeks Persepsi Korupsi Indonesia di tahun 2021 masih jauh dibawah negara Singapura (85), Malaysia (48) dan Vietnam (39).

*** Disclaimer : 

- Nilai Indeks Persepsi Korupsi (IPK) merupakan hasil survey yang dipublikasi secara tahunan oleh organisasi non-pemerintah berskala internasional yang bertekad untuk memerangi ketidakadilan yang disebabkan oleh korupsi.

- Survey tersebut mengurutkan 180 negara di dunia berdasarkan tingkat persepsi atau anggapan masyarakat mengenai korupsi yang terjadi pada jabatan publik dan politik. Indeks Persepsi Korupsi menggunakan skala 0 (korupsi yang tinggi) hingga 100 (korupsi yang rendah). Oleh karena itu, semakin tinggi nilai persepsi korupsi sebuah negara artinya semakin rendah pula korupsi yang terjadi di negara tersebut.
